## Poisson distribution?

$$p(k \mid \lambda) = \frac{\lambda^k e^{-\lambda}}{k!}$$

```python
import numpy as np
from math import factorial
from scipy.stats import poisson
import matplotlib.pyplot as plt

data = np.array([16, 24, 16,12, 16, 11, 14,15, 9, 14, 7])

lambda_ = np.mean(data)
# Check
lambda_

# range of values
k_values = np.arange(0, max(data)+1)

# Calculate the PMF values
pmf_values = poisson.pmf(k_values, lambda_)

# Plotting the PMF
plt.bar(k_values, pmf_values, color='blue', alpha=0.7)
plt.title('Poisson Distribution of Car Accidents During Rush Hour')
plt.xlabel('Number of Accidents')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

The assumption I made here is that the number of accidents follows a Poisson distribution because it counts the number of events that happen within a fixed interval of time. This means the probability of one accident occurring does not affect the probability of another event. Therefore, the mean and variance of the distribution are equal. In order to use Poisson, we have make sure each event is independently to eachother. 

Question 2: Finding the prameter

```python
def bayesian_norm(mu, sigma):
    """
    Calculating the posterior mean and standard deviation of multiple normal distributions
    :param mu: array of means
    :param sigma: array of standard deviations
    :return: mu, sigma
    """
    # Calculate the precision of the normal distribution
    precision = 1 / (sigma ** 2)
    
    # Calculate the posterior mean
    mu_posterior = np.sum(mu * precision) / np.sum(precision)

    # Calculate the posterior standard deviation
    sigma_posterior = 1 / np.sqrt(np.sum(precision))


    return mu_posterior, sigma_posterior
```

```python
mu = lambda_
sigma = np.array([2, 1, 2,3])

mu_posterior, sigma_posterior = bayesian_norm(mu, sigma)
print("Posterior Mean:", mu_posterior)
print("Posterior Standard Deviation:", sigma_posterior)
```
