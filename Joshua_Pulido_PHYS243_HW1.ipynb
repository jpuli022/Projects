Question 1: Part  A: reqriting the fib_rec code as an iteration

```python
# Professor's Fibanocci code.
def fib_rec(n):
    if n == 0:
        return 1
    elif n == 1:
        return 1
    else:
        return fib_rec(n-1)+fib_rec(n-2)
```

```python
fib_rec(3)
```

```python
# My fib Fibonacci sequence iterative implementation
def fib_iter(n):
    if n == 0:
        return 1
    elif n == 1:
        return 1
    else:
        print(0)

    a,b = 0,1
    for i in range(2,n+1):
        print(b)
        a, b = b, a + b
    return b
```

```python
# Fib_iter check. 
fib_iter(6)
```

```python


%%time
# Check CPU times of fibonacci iterative function
for n in range(100):
    fib_iter(10)
```

Part B: The iterative function is typically faster than the recursive version for larger values of n. This is because the recursive function computes the same Fibonacci number multiple times without storing which leads to a lack of memorization. This results leads to an exponential growth in computation time (O(2^n)). The iterative approach  computes each Fibonacci number exactly once, making it O(n) in time complexity.

Part C: 

```python
import time 


def timer(func, n, k):
    
    runtimes = []
    for _ in range(k):
        start_time = time.time()
        func(n)   
        end_time = time.time()
        runtime = end_time - start_time
        runtimes.append(runtime)

    average_time = sum(runtimes) / k
    return average_time
```

Part 4: plotting the fib_rec

```python
import matplotlib.pyplot as plt

n_values = range(10,21)

k = 10

times_recursive = [timer(fib_rec, n, k) for n in n_values]
times_iterative = [timer(fib_iter, n, k) for n in n_values]

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(n_values, times_recursive, label='Recursive Fibonacci', marker='o')
plt.plot(n_values, times_iterative, label='Iterative Fibonacci', marker='o')
plt.xlabel('n (Fibonacci Index)')
plt.ylabel('Average Time (seconds)')
plt.title('Comparative Execution Time of Fibonacci Functions')
plt.legend()
plt.grid(True)
plt.show()

```

Question 2: Part A. Linalg sequences
Given the Matrix below answer the questions: Find the determinant, transpose, inverse(if exist) for M.
M =(
1 −4 2
−4 1 −2
2 −2 −2)

```python
import numpy as np

M = np.array([[1,-4,2], [-4,1,-2], [2,-2,-2]])

# Determinant
det_M = np.linalg.det(M)

# Transpose 
# Nothing to transpose? 
transpose_M = M.T


# Inverse
inverse_M = np.linalg.inv(M) if np.linalg.det(M) != 0 else "Inverse does not exist"

print("Original:\n", M)
print("Determinant:", det_M)
print("Transpose:\n", transpose_M)
print("Inverse:\n", inverse_M)

```

Part 2: eigenvalues and eigenvectors

```python
from numpy.linalg import eig

w,v = eig(M)
print("Eigenvalues:", w)
print("Eigenvectors:", v)
```

Part 3: Finding the Gradient value for the equation.

```python
# import everything from sympy import *
from sympy import *
import sympy as sp

# Define symbols
x11, x12, x13, x21, x22, x23, x31, x32, x33 = sp.symbols('x11 x12 x13 x21 x22 x23 x31 x32 x33')

# Define the function
f =(x11**2*x22*x23) + (x11*x12*x13*x31) - (x33**2*x32*x21)

# Compute gradient
A = sp.Matrix([[x11, x12, x13], [x21, x22, x23], [x31, x32, x33]])

gradient = [sp.diff(f, var) for var in A]

print("Gradient:\n", Matrix(3, 3, gradient))

```
