```python
# Go into .env file and get the api_key from variable springer_api_key
api_key = os.getenv("springer_api_key")
```

To access a website:
* Request a Site
We need to give headers and parameters.
Headers tell the site who we are and parameters are specific arguments we request from the site and its API

```python
#Get User-Agent to supply for headers so the API we are accessing has more information on us
r = requests.get('https://httpbin.org/user-agent')
r
```

```python
#Load it into a Python data structure format
#json.loads ---> load-s ----> load string as JSON
ua = json.loads(r.text)['user-agent']
```

```python
# search website.com/robots.txt for rules on acceptable bots for a website.com
```

```python
headers = {"name":"Finn","email":"finn.mokrzycki@virginia.edu","User-Agent":ua}
params = {'api_key':api_key, 'q':'Molecular Compounds'}

base_url = 'https://api.springernature.com'
endpoint = '/openaccess/json'

r = requests.get(base_url + endpoint,headers=headers,params=params)
r #Shooting for 200 = Successful
```

```python
springer = json.loads(r.text)
```

```python
springer
```

```python
#Turn JSON into a dataframe
pd.json_normalize(springer)
```

```python
#If the JSON isnt looking good and you need to go down a nested layer
pd.json_normalize(springer,record_path=['records'])
```

```python
params = {'api_key':api_key, 'q':'doi:10.1007/s44174-024-00158-4'}

base_url = 'https://api.springernature.com'
endpoint = '/openaccess/jats' #Grab full text

r = requests.get(base_url + endpoint,headers=headers,params=params)
r #Shooting for 200 = Successful
```
